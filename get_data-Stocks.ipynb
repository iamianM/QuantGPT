```python
import pandas as pd
from zipfile import ZipFile
from os import listdir
from os.path import isfile, join
import datetime
import time
import os
import numpy as np
```

```python
t1 = time.time()
path = '../FinRL-Library/firstratedata'

seq_len = 300
frame = '1T'

zip_files = [f for f in listdir(path) if f.endswith('.zip')]
npy_files = [f for f in listdir('data/stocks') if f.endswith('.npy')]
for zip_file in zip_files:
    folder_name = zip_file.split('.')[0]
        
    print(folder_name)
    if folder_name + '.npy' in npy_files:
        continue
    zf = ZipFile(f'{path}/{zip_file}') 
    data_files = zf.filelist
    data = []
    for file in data_files:
        df = pd.read_csv(zf.open(file), names=['time', 'Open', 'High', 'Low', 'close', 'Volume'])
        
        df = df[['time', 'close']]
        df['time'] = df['time'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
        df = df.set_index('time')
        df = df[~df.index.duplicated()]
        if df.shape[0] == 0:
            continue
        print(file.filename)
        df = df.resample(frame).fillna(None)

        temp = (~df.isna()).cumsum()
        df['index'] = np.concatenate((((temp[seq_len:].values - temp[:-seq_len].values) >= [0.9*seq_len])[:,0], [False]*seq_len))

        df = df.fillna(method='ffill')
        close_log = 100*(df.iloc[1:]['close'].values - df.iloc[:-1]['close'].values)/df.iloc[:-1]['close'].values
        close_log = np.concatenate(([None], np.sign(close_log)*np.log(abs(close_log)+1)))
        df['close_log'] = close_log
        df = df.iloc[1:]

        df = df.reset_index()
        idxs = df[df['index']].index
        prev_idxs = []
        count = 0
        for i in idxs:
            if i+2 in prev_idxs:
                continue
            seq = df.loc[i+2:i+seq_len]
            data.append(seq['close_log'].values)
            prev_idxs = list(range(i+2, i+seq_len))
            count += 1
        print(count*seq_len, df['index'].sum(), len(df), len(data))
        
    np.save(f'data/stocks/{folder_name}', data)
    
print(time.time() - t1)
```
