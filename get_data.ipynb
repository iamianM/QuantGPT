```python
seq_len = 300
directory = '../FX-1-Minute-Data/output'
folders = [x[0].split('\\')[-1] for x in os.walk(directory)][1:]
data = []
for folder in folders:
    zip_files = [f for f in listdir(f'{directory}/{folder}') if f.endswith('.zip') and '2021' in f]

    for file in zip_files:
        print(file)
        zip_file = ZipFile(f'{directory}/{folder}/{file}')
        df = pd.read_csv(zip_file.open(file[:-4]+'.csv'), names=['datetime', 'open', 'high', 'low', 'close', 'volume'], sep=';')
        df = df[['datetime', 'close']]
        df['datetime'] = df['datetime'].apply(lambda x: datetime.datetime.strptime(x, '%Y%m%d %H%M%S'))
        df = df.set_index('datetime')
        df = df[~df.index.duplicated()]
        df = df.resample('1T').fillna(None)
        
        temp = (~df.isna()).cumsum()
        df['index'] = np.concatenate((((temp[seq_len:].values - temp[:-seq_len].values) >= [0.9*seq_len])[:,0], [False]*seq_len))
        
        df = df.fillna(method='ffill')
        close_log = 100*50*(df.iloc[1:]['close'].values - df.iloc[:-1]['close'].values)/df.iloc[:-1]['close'].values
        close_log = np.concatenate(([None], np.sign(close_log)*np.log(abs(close_log)+1)))
        df['close_log'] = close_log
        df = df.iloc[1:]
        
        df = df.reset_index()
        idxs = df[df['index']].index
        prev_idxs = []
        count = 0
        for i in idxs:
            if i+2 in prev_idxs:
                continue
            seq = df.loc[i+2:i+seq_len]
            data.append(seq['close_log'].values)
            prev_idxs = list(range(i+2, i+seq_len))
            count += 1
        print(count*seq_len, df['index'].sum(), len(df), len(data))
```
